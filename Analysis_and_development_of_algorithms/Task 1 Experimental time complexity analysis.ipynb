```python
# Download all the necessary libraries
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt 
import math
import timeit 
```

```python
# Set the sizes of vectors
sizes = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
```

```python
# We set the function so that it draws a graph for us
def graph(times, sizes, title):
    plt.plot([min(sizes),max(times[0])], [min(times[1]),max(times[1])], color='orange', linewidth=3, alpha = 0.5, label = 'O(n)')
    plt.scatter(times[0], times[1], color='blue', label = 'empirical time complexity ')
    plt.xlabel('Number of elements')
    plt.ylabel('Time')
    plt.grid(c='#DCDCDC')
    plt.title(title)
    #If we need to draw in the logarithmic scale - plt.xscale('log') 
    plt.legend(loc = 'best')
    plt.show()
```

## 1) Constant function

```python
# Implement the function
def constant_f(v):
    return(500)

times_1 = [[],[]]
# For each of 10 samples of size elements
for size in sizes:
    v = np.random.uniform(1,500, size)
    tests_results = []
    # We carry out five experiments
    for i in range(0,5):
        start = timeit.default_timer() # Start to time
        constant_f(v)
        tests_results.append(timeit.default_timer()-start) # Finish and write the result to the list
        
    mean_time = np.mean(np.array(tests_results)) # Average time from the list
    print('Test time: ', mean_time)
    times_1[0].append(size)
    times_1[1].append(mean_time)

# Draw a graph, here we add another line that shows the complexity of O(1)
plt.plot([min(sizes),max(times_1[0])], [np.mean(times_1[1]),np.mean(times_1[1])], color='green', linewidth=3, alpha = 0.5, label = 'O(1)')
graph(times_1, sizes, 'The dependence of the running time of the constant \nalgorithm on the number of elements in the vector') 
```

So there is the O(1) time compexity for our constant algorithm becuase of points of empirical time complexity scattered evenly relative to the green line without any dependence on number of elements.

## 2) The sum of elements

```python
# Can use numpy function
np.sum(v)
```

```python
# Implement the function
def calc_f(v):
    summa = 0
    for i in v:
        summa += i
    return(summa)

times_2 = [[],[]]
# For each of 10 samples of size elements
for size in sizes:
    v = np.random.uniform(1,500, size)
    tests_results = []
    # We carry out five experiments
    for i in range(0,5):
        start = timeit.default_timer() # Start to time
        calc_f(v)
        tests_results.append(timeit.default_timer()-start) # Finish and write the result to the list
        
    mean_time = np.mean(np.array(tests_results)) # Average time from the list
    print('Test time: ', mean_time)
    times_2[0].append(size)
    times_2[1].append(mean_time)
    
# Draw a graph
graph(times_2, sizes, 'The dependence of the running time of the sum \nalgorithm on the number of elements in the vector')
```

The time complexity of the algorithm is O(n).

## 5) Polinomial

```python
# let's define a function for calculating in two ways
def polinomial_1(v, x = 1.5):
    Px = []
    for i in range(0, len(v)):
        P = v[i]*x**i
        Px.append(P)
    return(sum(Px))

# By Horner's method
def polinomial_2(v, x = 1.5):
    Px = 0
    for i in range(len(v)-1, -1, -1):
        Px = Px*x + v[i]
    return(Px)

times_51 = [[],[]]
# For each of 10 samples of size elements
for size in sizes:
    v = np.random.uniform(1,500, size)
    tests_results = []
    # We carry out five experiments
    for i in range(0,5):
        start = timeit.default_timer() # Start to time
        polinomial_1(v)
        tests_results.append(timeit.default_timer()-start) # Finish and write the result to the list
        
    mean_time = np.mean(np.array(tests_results)) # Average time from the list
    print('Test time: ', mean_time)
    times_51[0].append(size)
    times_51[1].append(mean_time)
    
# Draw a graph
graph(times_51, sizes,
      'The dependence of the running time of the calculating P(x) \nalgorithm on the number of elements in the vector')

times_52 = [[],[]]
# For each of 10 samples of size elements
for size in sizes:
    v = np.random.uniform(1,500, size)
    tests_results = []
    # We carry out five experiments
    for i in range(0,5):
        start = timeit.default_timer() # Start to time
        polinomial_2(v)
        tests_results.append(timeit.default_timer()-start) # Finish and write the result to the list
        
    mean_time = np.mean(np.array(tests_results)) # Average time from the list
    print('Test time: ', mean_time)
    times_52[0].append(size)
    times_52[1].append(mean_time)

graph(times_52, sizes,
      "The dependence of the running time of the calculating P(x) by Horner's method \nalgorithm on the number of elements in the vector")
```

The time complexity of the algorithms is O(n).

## 6) The bubble sort 

```python
def bubble_sort(v):
    for i in range(0, len(v)):
        for j in range(0, len(v)-i-1):
            if v[j] > v[j+1]:
                v[j], v[j+1] = v[j+1], v[j]
            else:
                pass
    return(v)

times_6 = [[],[]]
# For each of 10 samples of size elements
for size in sizes:
    v = np.random.uniform(1,500, size)
    tests_results = []
    # We carry out five experiments
    for i in range(0,5):
        start = timeit.default_timer() # Start to time
        bubble_sort(v)
        tests_results.append(timeit.default_timer()-start) # Finish and write the result to the list
        
    mean_time = np.mean(np.array(tests_results)) # Average time from the list
    print('Test time: ', mean_time)
    times_6[0].append(size)
    times_6[1].append(mean_time)

# Draw a plot   
data1 = times_6[1]
data2 = np.array(sizes)**2

fig, ax1 = plt.subplots()
ax1.set_xlabel('Number of elements')
ax1.set_ylabel('Time')
ax1.scatter(sizes, data1, color='blue')
ax1.tick_params(axis='y')
plt.grid(c='#DCDCDC')
plt.ylim(0, 0.17)

ax2 = ax1.twinx()  
ax2.plot(sizes, data2, color='red', linewidth=3, alpha = 0.5, label = 'O(n^2)')
ax2.tick_params(axis='y')
ax2.axis('off')
plt.title("The dependence of the running time of bubble sort \nalgorithm on the number of elements in the vector")
plt.legend(loc = 'lower right')
plt.show()
```

The time complexity of the algorithm is O(n^2).

# Matrix product for A and B

```python
# We set the sizes of the matrices, 50 by 50, 100 by 100, etc.
sizes = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]

times_7 = [[],[]]
# For each of 10 selections of size elements
for n in sizes:
    # Generate a matrix A and B of size n by n elements
    A = np.random.random((n, n))
    B = np.random.random((n, n))
    tests_results = []
    # We carry out five experiments
    for i in range(0,5):
        start = timeit.default_timer() # Start to time
        np.dot(A,B) # Using the numpy function
        tests_results.append(timeit.default_timer()-start) # Finish and write the result to the list
        
    mean_time = np.mean(np.array(tests_results)) # Average time from the list
    print('Test time: ', mean_time)
    times_7[0].append(n)
    times_7[1].append(mean_time)

# Draw a plot   
data1 = times_7[1]
data2 = np.array(sizes)**3

fig, ax1 = plt.subplots()
ax1.set_xlabel('Number of elements')
ax1.set_ylabel('Time')
ax1.scatter(sizes, data1, color='blue')
ax1.tick_params(axis='y')
plt.grid(c='#DCDCDC')

ax2 = ax1.twinx()  
ax2.plot(sizes, data2, color='red', linewidth=3, alpha = 0.5, label = 'O(n^3)')
ax2.tick_params(axis='y')
ax2.axis('off')
plt.title("The dependence of the running time of np.dot(A,B) \nalgorithm on the number of n")
plt.legend(loc = 'lower right')
plt.show()
```

As can be seen from the graph, with an increase in the number of elements in the matrix, the complexity of the calculations increases too. It is worth noting that the running time of the algorithm depends on n nonlinearly, but is most likely approximated by the dependence O(n^3).
